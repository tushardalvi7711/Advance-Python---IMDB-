```python
# making new dataframe for analysis
new_df = pd.DataFrame({'imdbid': df['imdbid'],
                       'title': df['title'],'akas': df['akas'].str.replace('\),', ');')})
```

```python
new_df
```

```python
# dropping the null values 
new_df.dropna(inplace = True)
```

```python
# splitting each title in the akas list
new_df2 = pd.DataFrame({'imdbid': new_df['imdbid'],
                        'title': new_df['title'],'akas': new_df['akas'].str.split('; ')})
```

```python
new_df2
```

```python
#aking the values of region from paranthesis and making new seperate columns by expanding
akas_data = new_df2.explode('akas')
akas_data[['akas', 'region']] = akas_data.akas.str.split('\((.*)\)',expand = True).iloc[:,[0,1]]
```

```python
akas_data
```

```python
akas_data['region'] = akas_data['region'].str.strip('()')
```

```python
#to find the unique regions 
regions = akas_data['region'].unique()
```

```python
regions
```

```python
def movie_regions_title(m,n):
    title_list = ((akas_data.groupby('title')).get_group(m)['region']).to_string()
    if n in title_list:
        result_df1 = akas_data['title'] == m
        result_df2 = akas_data['region'] == n
        final_result = akas_data.loc[((result_df1) & (result_df2)),['imdbid', 
                                                                    'title', 
                                                                    'akas',
                                                                    'region']]
        print(final_result)

    else:
        result_df1 = akas_data['title'] == m
        final_result = akas_data.loc[(result_df1),['imdbid', 'title', 'akas',
                                                                    'region']]
        print(final_result)

movie_regions_title('You Were Never Really Here', 'abc')
```
